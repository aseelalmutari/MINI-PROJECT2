```python
import numpy as np
X=np.array([[1.,0.5],[2.,-1.],[0.,3.]])
W=np.array([[0.2,-0.1,0.5],[0.7,0.3,-0.2]])
b=np.array([0.1,0.0,-0.3])
print(X.shape)
print(W.shape)
print(b.shape)
means=X.mean(axis=0,keepdims=True)
stds=X.std(axis=0,keepdims=True)
X_std=(X-means)/stds
print(means.shape)
print(stds.shape)
print(X_std.shape)
Y=X_std@W+b
print(Y.shape)
Y_unit=Y/np.linalg.norm(Y,axis=1,keepdims=True)
print(Y_unit.shape)
```
